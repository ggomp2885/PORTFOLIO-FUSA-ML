### General Imports & Pandas Settings

```python
import pandas as pd
import os


# pd.set_option("max_colwidth", 250)
pd.set_option("max_colwidth", 10)
pd.set_option("display.max_columns", 60)
pd.set_option('max_rows', None)
```

### 1. Loading in 5 years of SEC data

```python
path = r"C:\Users\ggomp\OneDrive\Documents\GitHub\Funders-USA-ML-Backend\datasets\datasets_SEC"
path, dirs, files = next(os.walk(path))
dirs_paths = ["{}\\".format(path) + i for i in dirs]


display(dirs_paths)
```

```python
# Working complete loop of excel merging

curr_qt_df = pd.DataFrame()
curr_df = pd.DataFrame()
df = pd.DataFrame()
list_of_docs = ('DISCLOSURE.tsv', 'INFORMATION.tsv', 'SUBMISSION.tsv')

for i in dirs_paths:
    curr_path = i
    #display(curr_path)
    path, dirs, files = next(os.walk(curr_path))
    #display(files)
    files_paths = ["{}\\".format(path) + i for i in files]  
    #display(files_paths)
    for i in files_paths:
        if i.endswith(list_of_docs):
            curr_df = pd.DataFrame()
            curr_df = pd.read_table(i)
            #display(curr_df.head())
            curr_qt_df = pd.concat([curr_df, curr_qt_df], axis=1)
            #print(curr_qt_df.shape)
        #print("if statement complete")
    #print("appending qt_df to df")
    df = df.append(curr_qt_df, ignore_index=True)
    print(df.shape)
    #print("resetting qt_df to empty")
    curr_qt_df = pd.DataFrame()
    #print("exit curr_path")
#print("exit entire loop")
print()
print("Master dataset")
display(df.head(2))
```

### 2. Merging all 5 years of data into one dataframe

```python

list_of_docs = ('DISCLOSURE.tsv', 'INFORMATION.tsv', 'SUBMISSION.tsv')
curr_qt_df = pd.DataFrame()
curr_df = pd.DataFrame()
df = pd.DataFrame()


for i in dirs_paths:
    curr_path = i
    path, dirs, files = next(os.walk(curr_path))
    files_paths = ["{}\\".format(path) + i for i in files] 
    
    for i in files_paths:
        if i.endswith(list_of_docs):
            curr_df = pd.DataFrame()
            curr_df = pd.read_table(i)
            curr_qt_df = pd.concat([curr_df, curr_qt_df], axis=1)
            
    df = df.append(curr_qt_df, ignore_index=True)
    print(df.shape)
    curr_qt_df = pd.DataFrame()
    
```
