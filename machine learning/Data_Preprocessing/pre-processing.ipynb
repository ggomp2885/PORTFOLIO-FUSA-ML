#### 1. General Imports & Pandas Settings

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import datetime
import os


# pd.set_option("max_colwidth", 250)
pd.set_option("max_colwidth", 150)
pd.set_option("display.max_columns", 60)
pd.set_option('max_rows', None)
```

#### 2. Loading master dataframe from saved Excel

```python
path = "/root/Funders-USA-ML-Backend/datasets/datasets_SEC/Aggregated SEC Data.xlsx"
df = pd.read_excel(path)


df = df.sort_values('FILING_DATE')
df.reset_index(drop=True, inplace=True)


df['FILING_DATE'] = pd.to_datetime(df['FILING_DATE'], format='%Y%m%d')
display(df.shape)
display(df.tail(2))
```

#### 3. Shrinking dataframe to MVP columns

```python
list_of_cols_MVP = ['CIK', 'FILING_DATE', 'SUBMISSION_TYPE', 'OFFERINGAMOUNT', 'PROGRESSUPDATE', 'NATUREOFAMENDMENT', 'MAXIMUMOFFERINGAMOUNT', 'CURRENTEMPLOYEES', 'TOTALASSETMOSTRECENTFISCALYEAR',
                       'TOTALASSETPRIORFISCALYEAR', 'CASHEQUIMOSTRECENTFISCALYEAR', 'CASHEQUIPRIORFISCALYEAR',
                       'ACTRECEIVEDRECENTFISCALYEAR', 'ACTRECEIVEDPRIORFISCALYEAR', 'SHORTTERMDEBTMRECENTFISCALYEAR',
                       'SHORTTERMDEBTPRIORFISCALYEAR', 'LONGTERMDEBTRECENTFISCALYEAR', 'LONGTERMDEBTPRIORFISCALYEAR',
                       'NETINCOMEMOSTRECENTFISCALYEAR', 'NETINCOMEPRIORFISCALYEAR', 'Form_C_Number', 'Funding_Success']
df = df[list_of_cols_MVP]


display(df.shape)
display(df.head(4))
display(df.tail(1))
```

#### 4. Add target column to master df

```python
df_length = len(df)
display(df_length)


df = df.assign(Form_C_Number = [0]*df_length)
df = df.assign(Funding_Success = [0]*df_length)
display(df.tail(2))
```

#### 5. Setting up custom loops to extract the needed information

##### Loop for creating list of Unique CIKs

```python
u_CIK_list = df['CIK'].unique()
display(u_CIK_list[0:5])
```

##### Loop for creating list containing the Progress update counts for each CIK

```python
xdf = df.copy()
PU_count = 0
PU_count_list = []


for i in u_CIK_list:
    
    for row, column in xdf[xdf['CIK'] == i].iterrows():
        if column[4] is not np.nan:
            PU_count += 1

    PU_count_list.append(PU_count)
    PU_count = 0
    
    
print(len(PU_count_list))
```

### 6. Loop for assignment of target feature (Funding Success)

```python
xdf = df.copy()
ndf = pd.DataFrame().reindex_like(xdf)
ndf = ndf.dropna()
display(xdf.head(2))


C_row_list = []
C_row_success_list = []
C_count = 0
C_cutoff_date = datetime.datetime.strptime('2019-07-01 00:00:00', '%Y-%m-%d %H:%M:%S')


    
for idx, val in enumerate(u_CIK_list):
    if PU_count_list[idx] > 0:        
        
        
        for row, column in xdf[xdf['CIK'] == val].iterrows():
            if column[2] == 'C' and column[1] < C_cutoff_date:
                C_row_success_list = []
                C_row_list.append(row)
                C_row_success_list.append(row)
                C_count +=1
                xdf.iloc[row, -2] = C_count
            
            
            if column[2] == 'C-W' and len(C_row_success_list) > 0:
                C_row_success_list.remove(C_row_list[-1])

                
            if column[4] is not np.nan:
                xdf.iloc[C_row_success_list, -1] = 1        
    
    
    
        for row, column in xdf[xdf['CIK'] == val].iterrows():

            
            if column[2] == 'C' and column[1] < C_cutoff_date:
                C_row_success_list = []
                C_row_list.append(row)
                C_row_success_list.append(row)
                C_count +=1
                xdf.iloc[row, -2] = C_count
            
            
            if column[2] == 'C-W' and len(C_row_success_list) > 0:
                C_row_success_list.remove(C_row_list[-1])
                
                
            if column[2] == 'C-AR':
                xdf.iloc[C_row_success_list, -1] = 1

                
    ndf = ndf.append(xdf.iloc[C_row_list,:])
    C_row_list = []
    C_row_success_list = []
    C_count = 0

    
print('loop complete')
ndf = ndf.astype({'CIK': int, 'Funding_Success': int, 'Form_C_Number': int})
display('Crafted Dataset below')
display(ndf.head(10))
```

### Final custom preprocessing before going into Data Wrangler

```python
display(ndf.tail(7))
nndf = ndf.copy()
display(nndf.tail(7))

# removing entries with offering amounts less than 9k
nndf = nndf[nndf['OFFERINGAMOUNT'] > 9000]
display(len(nndf))

#droping NA values
nndf = nndf.dropna(how='any', subset =["OFFERINGAMOUNT"])
display(len(nndf))

#dropping NA values
nndf = nndf.dropna(how='any', subset =['MAXIMUMOFFERINGAMOUNT'])
display(len(nndf))
display(len(nndf[nndf['CURRENTEMPLOYEES'] > 40]))
display(nndf.tail(7))

# calculating class balance %
pos_cases = nndf[nndf['Funding_Success'] == 1]
total = len(nndf)
success = len(pos_cases)
no_funding = total - success

# printing class balance %
print('total amount of C forms within date cutoff: {}, successes {}, no funding {}'.format(total, success, no_funding))
print(success / total *100)

display(nndf.describe())
```
